Iris data set analysis

In the data set provided... There are three species of Iris flower mentioned which are setosa, versicolor and virginica. The above mentioned species can be differentiated on the basis of sepal length, sepal width, petal length and petal width.

```python
### importing required libraries

import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
### uploading iris data files

Iris = pd.read_csv("Iris.csv")
Iris
```

There are 50 values given for each type of species. There are 4 columns and the Output line 48 shows the mean, standard deviation, minimum, maximum. 


```python
### Display the plot

Iris.plot(kind = "scatter", x ="SepalLengthCm", y = "SepalWidthCm", color = "blue", s=60)
plt.show()
```

```python
### modifying the graph above by assigning different colours to each species

sns.FacetGrid(Iris, hue="Species", size=10)\
.map(plt.scatter, "SepalLengthCm" , "SepalWidthCm") \
.add_legend()
plt.show()
```

```python
### Box plot

Iris.plot(kind = "box",subplots=True, layout=(3,2),figsize=(8,12))
plt.show()
```

```python
## Visualisation using box plot

sns.boxplot(x='Species', y="PetalLengthCm", data= Iris)
plt.show()
```

```python
sns.stripplot(x='Species', y="PetalLengthCm", data= Iris, jitter=True, edgecolor= "gray")
```

```python
## Voilin plot

sns.violinplot(x='Species',y ="PetalLengthCm",palette={"blue","red","yellow"},data=Iris, size=10)
```

The violinplot shows density of the length and width in the species. The thinner part denotes that there is less density whereas the fatter part conveys higher density

```python
sns.pairplot(Iris.drop("Id", axis=1), hue="Species", size=3)
plt.show()
```

```python
### Andrews plot

plt.figure(figsize=(12,6))
pd.plotting.andrews_curves(Iris.drop("Id",axis=1), "Species", colormap='rainbow')
plt.show()
```
